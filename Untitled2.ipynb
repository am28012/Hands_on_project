```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
df = pd.read_csv('/content/BostonHousing.csv')
```

```python
plt.figure(figsize=(6, 5))
sns.boxplot(y='MEDV', data=df)
plt.title('Boxplot of Median Value of Owner-Occupied Homes (MEDV)')
plt.ylabel('MEDV ($1000s)')
plt.grid(True)
plt.show()
```

```python
# Provide a boxplot for the medv variable vs the AGE variable. (Discretize the age variable into three groups of 35 years and younger, between 35 and 70 years and 70 years and older)

# Discretize the AGE variable into three groups
df['AGE_GROUP'] = pd.cut(df['age'], bins=[0, 35, 70, df['age'].max()], labels=['<=35', '35-70', '>70'])

# Create the boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='AGE_GROUP', y='medv', data=df)
plt.title('Boxplot of MEDV vs. AGE Group')
plt.xlabel('Age Group')
plt.ylabel('MEDV ($1000s)')
plt.grid(axis='y')
plt.show()
```

```python
#  Provide a scatter plot to show the relationship between Nitric oxide concentrations and the proportion of non-retail business acres per town. What can you say about the relationship?

# Provide a scatter plot to show the relationship between Nitric oxide concentrations and the proportion of non-retail business acres per town.
plt.figure(figsize=(8, 6))
sns.scatterplot(x='nox', y='indus', data=df)
plt.title('Scatter Plot of Nitric Oxide Concentration vs. Non-Retail Business Acres')
plt.xlabel('Nitric Oxide Concentration (NOX)')
plt.ylabel('Proportion of Non-Retail Business Acres (INDUS)')
plt.grid(True)
plt.show()

# Relationship analysis
print("Analysis of the relationship between Nitric oxide concentrations (NOX) and the proportion of non-retail business acres (INDUS):")
print("The scatter plot shows a generally positive correlation between NOX and INDUS.")
print("As the proportion of non-retail business acres increases, the concentration of nitric oxides also tends to increase.")
print("This suggests that areas with more industrial and non-retail business activity tend to have higher levels of air pollution as measured by nitric oxide.")
print("The relationship appears somewhat linear, although there might be some outliers or variations.")
```

```python
#  Use α = 0.05
# Perform the test Statistics.
# State the conclusion from the test.    Is there a significant difference in median value of houses bounded by the Charles river or not? (T-test for independent samples)
# Is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)? (ANOVA)
# Can we conclude that there is no relationship between Nitric oxide concentrations and proportion of non-retail business acres per town? (Pearson Correlation)
# What is the impact of an additional weighted distance  to the five Boston employment centres on the median value of owner occupied homes? (Regression analysis)

alpha = 0.05

# T-test for independent samples (Charles River)
# State the null and alternative hypotheses
# H0: The median value of houses bounded by the Charles river is not significantly different from those not bounded by the Charles river.
# H1: The median value of houses bounded by the Charles river is significantly different from those not bounded by the Charles river.
charles_river_houses = df[df['chas'] == 1]['medv']
other_houses = df[df['chas'] == 0]['medv']
t_statistic, p_value = scipy.stats.ttest_ind(charles_river_houses, other_houses)
print(f"T-test for MEDV vs. Charles River (CHAS):")
print(f"  Test Statistic: {t_statistic:.4f}")
print(f"  P-value: {p_value:.4f}")

# Conclusion for T-test
if p_value < alpha:
  print("  Conclusion: Since the p-value is less than alpha (0.05), we reject the null hypothesis.")
  print("  There is a significant difference in the median value of houses bounded by the Charles river compared to those not bounded.")
else:
  print("  Conclusion: Since the p-value is greater than alpha (0.05), we fail to reject the null hypothesis.")
  print("  There is no significant difference in the median value of houses bounded by the Charles river compared to those not bounded.")

print("-" * 50)

# ANOVA (MEDV vs. AGE_GROUP)
# State the null and alternative hypotheses
# H0: The mean median value of houses is the same for all age groups.
# H1: The mean median value of houses is different for at least one age group.
model = ols('medv ~ C(AGE_GROUP)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print("ANOVA for MEDV vs. AGE Group:")
print(anova_table)
p_value_anova = anova_table['PR(>F)'][0]

# Conclusion for ANOVA
if p_value_anova < alpha:
  print("  Conclusion: Since the p-value is less than alpha (0.05), we reject the null hypothesis.")
  print("  There is a significant difference in the Median values of houses (MEDV) for at least one proportion of owner occupied units built prior to 1940 (AGE) group.")
else:
  print("  Conclusion: Since the p-value is greater than alpha (0.05), we fail to reject the null hypothesis.")
  print("  There is no significant difference in the Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE) group.")

print("-" * 50)

# Pearson Correlation (NOX vs. INDUS)
# State the null and alternative hypotheses
# H0: There is no linear relationship between Nitric oxide concentrations and the proportion of non-retail business acres per town.
# H1: There is a linear relationship between Nitric oxide concentrations and the proportion of non-retail business acres per town.
correlation_coefficient, p_value_corr = scipy.stats.pearsonr(df['nox'], df['indus'])
print(f"Pearson Correlation between NOX and INDUS:")
print(f"  Correlation Coefficient: {correlation_coefficient:.4f}")
print(f"  P-value: {p_value_corr:.4f}")

# Conclusion for Pearson Correlation
if p_value_corr < alpha:
  print("  Conclusion: Since the p-value is less than alpha (0.05), we reject the null hypothesis.")
  print("  We conclude that there is a significant linear relationship between Nitric oxide concentrations and the proportion of non-retail business acres per town.")
else:
  print("  Conclusion: Since the p-value is greater than alpha (0.05), we fail to reject the null hypothesis.")
  print("  We cannot conclude that there is a significant linear relationship between Nitric oxide concentrations and the proportion of non-retail business acres per town.")

print("-" * 50)

# Regression analysis (MEDV vs. DIS)
# State the model
# MEDV = beta_0 + beta_1 * DIS + epsilon
X = df['dis']
y = df['medv']
X = sm.add_constant(X) # Add an intercept term to the model
model = sm.OLS(y, X).fit()
print("Regression Analysis for MEDV vs. Weighted Distance to Employment Centres (DIS):")
print(model.summary())

# Interpretation of the impact of DIS on MEDV
print("\nInterpretation of the impact of DIS on MEDV:")
print(f"  The coefficient for 'dis' represents the estimated change in MEDV for a one-unit increase in the weighted distance to five Boston employment centres.")
print(f"  From the regression output:")
print(f"  Coefficient for DIS: {model.params['dis']:.4f}")
print(f"  P-value for DIS: {model.pvalues['dis']:.4f}")
print(f"  R-squared: {model.rsquared:.4f}")

if model.pvalues['dis'] < alpha:
  print("  Conclusion: Since the p-value for DIS is less than alpha (0.05), the weighted distance to employment centres (DIS) has a statistically significant impact on the median value of owner occupied homes (MEDV).")
  print(f"  For each additional weighted distance unit, the median value of owner occupied homes is estimated to change by approximately ${model.params['dis']:.2f} thousand (given the coefficient).")
  if model.params['dis'] > 0:
    print("  The positive coefficient indicates that as the weighted distance increases, the median value of homes tends to increase.")
  else:
     print("  The negative coefficient indicates that as the weighted distance increases, the median value of homes tends to decrease.")
else:
  print("  Conclusion: Since the p-value for DIS is greater than alpha (0.05), the weighted distance to employment centres (DIS) does not have a statistically significant impact on the median value of owner occupied homes (MEDV) at the 0.05 significance level.")
```
